```python
import pandas as pd
import numpy as np
import sklearn
import os
import math
df= pd.read_csv("res_dep8.csv",encoding='utf8')
#Remove the null value of Star
for i, row in df.iterrows():
    if math.isnan(row['Star']):
        df.drop(i, inplace=True)    
y_star =df['Star']
for i in range(len(y_star)):
    if y_star[i] >=4:
        df['label'][i]= 1
    elif y_star[i] <=2:
        df['label'][i]= 0
df=df.drop(df[(df['Star']==3)].index,axis=0)
df.head(30)
```

```python
df.dropna(inplace = True)
df.shape
```

```python
import pandas as pd
import numpy as np
np.random.seed(123)
from sklearn.model_selection import train_test_split

df_X=df['Comment']
df_y=df['label']
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, 
                                                    test_size = 0.2,
                                                    random_state=123)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# instantiate the vectorizer. We set the vocabulary size to 50000
num_features = 40000 
vect = CountVectorizer(min_df=4,max_df=0.8,max_features=num_features)
# fit on data and transform data to vector (document-term matrix)
X_train_dtm = vect.fit_transform(X_train)
X_test_dtm = vect.transform(X_test)
```

```python
X_train_dtm.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(min_df=3,max_df=0.8,use_idf=True,ngram_range=(1,3)) 
# fit on data and transform data to vector (document-term matrix)
X_train_dtm1 = tfidf_vectorizer.fit_transform(X_train)
X_test_dtm1 = tfidf_vectorizer.transform(X_test)
print(X_train_dtm1.shape)
terms = tfidf_vectorizer.get_feature_names()
print(terms)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
# instantiate a logistic regression model
logreg = LogisticRegression(penalty='l2',C=0.1)

# train the model using X_train_dtm
logreg.fit(X_train_dtm, y_train)

# used the test label to return the result of predication either 0 or 1.
y_pred_class = logreg.predict(X_test_dtm)
metrics.accuracy_score(y_test, y_pred_class)
```

```python
from sklearn.model_selection import cross_val_score
precision=cross_val_score(logreg, X_test_dtm, y_test,cv=10,scoring='precision')
f1_score=cross_val_score(logreg,X_test_dtm, y_test,cv=10,scoring='f1')
recall=cross_val_score(logreg, X_test_dtm, y_test,cv=10,scoring='recall')

print('Precision',np.mean(precision),precision)
print('Recall',np.mean(recall),recall)
print('F1_Score',np.mean(f1_score),f1_score)
```

```python
from sklearn.metrics import confusion_matrix 
import matplotlib.pyplot as plt 
import numpy as np
def plot_confusion_matrix(cm, labels_name, title):
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]    # 归一化
    plt.imshow(cm, interpolation='nearest',cmap=plt.cm.Blues)    # 在特定的窗口上显示图像
    plt.title(title)    # 图像标题
    plt.colorbar()
    num_local = np.array(range(len(labels_name)))    
    plt.xticks(num_local, labels_name, rotation=90)    # 将标签印在x轴坐标上
    plt.yticks(num_local, labels_name)    # 将标签印在y轴坐标上
    plt.ylabel('True label')    
    plt.xlabel('Predicted label')
    
cm = confusion_matrix(y_test, y_pred_class)
print(cm)
labels_name=['positive','negative']

plot_confusion_matrix(cm, labels_name, "Confusion Matrix")
plt.show()
```
